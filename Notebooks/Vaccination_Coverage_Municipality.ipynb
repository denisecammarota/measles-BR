## Vaccination Coverage by Municipality (FIRST AND SECOND DOSE)

```python
df_1 = pd.read_csv('../Data/CV_TP_1_MUN.csv', encoding='cp1252', delimiter = ';', decimal = ',') 
df_2 = pd.read_csv('../Data/CV_TP_2_MUN.csv', encoding='cp1252', delimiter = ';', decimal = ',') 
```

```python
df_1 = df_1.drop(df_1.tail(1).index)
df_2 = df_2.drop(df_2.tail(1).index)
```

```python
df_1['ID_MUNICIP'] = df_1['Município'].str[:6].astype(int)
df_2['ID_MUNICIP'] = df_2['Município'].str[:6].astype(int)
```

```python
list_muns = [130250,130260,351630,140045]
names_muns = ['Manacapuru','Manaus','Francisco Morato','Pacaraima']
```

```python
df_1 = df_1[df_1['ID_MUNICIP'].isin(list_muns)]
df_2 = df_2[df_2['ID_MUNICIP'].isin(list_muns)]
```

```python
df_1 = df_1[['ID_MUNICIP','1999', '2000', '2001', '2002', '2003', '2004', '2005',
       '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014',
       '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']]
df_2 = df_2[['ID_MUNICIP', '2013', '2014',
       '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']]
```

```python
df_1 = df_1.set_index('ID_MUNICIP')
df_2 = df_2.set_index('ID_MUNICIP')
```

```python
plt.plot(df_1.T,'-o', label = names_muns)
plt.title('Vaccination Coverage of First Dose (1999-2022)')
plt.xticks(rotation = 45)
plt.legend()
plt.show()
```

```python
plt.plot(df_2.T,'-o', label = names_muns)
plt.title('Vaccination Coverage of Second Dose (2013-2022)')
plt.xticks(rotation = 45)
plt.legend()
plt.show()
```
