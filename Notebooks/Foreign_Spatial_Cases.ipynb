```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
import geopandas as gpd
import geobr
```

```python
sys.path.append('..')
```

```python
df_inf = pd.read_csv('../Data/analyzed/inf_abroad.csv')
df_res = pd.read_csv('../Data/analyzed/res_abroad.csv')
mun_file = pd.read_csv('../Data/municipalities.csv')
```

```python
mun_file['codigo'] = mun_file['municipio'].astype(str).str[:6].astype(int)
mun_file = mun_file[['municipio','name','codigo','pop_21']]
```

```python
df_inf = df_inf.merge(mun_file, how = 'left', left_on = 'ID_MUNICIP', right_on = 'codigo')
df_res = df_res.merge(mun_file, how = 'left', left_on = 'ID_MUNICIP', right_on = 'codigo')
```

```python
map_br = geobr.read_municipality()
```

```python
    map_br_aux = map_br.copy()
    map_br_aux = map_br_aux.merge(df_res, how = 'left', left_on = 'code_muni', right_on = 'municipio')
    fig, axs = plt.subplots(figsize = (12,12))
    map_br_aux = map_br_aux.fillna(0)
    map_br_aux.plot(
        ax=axs, # Axis de destino do gráfico
        column="CASO", # Coluna com os valores usados para colorir o mapa
    )
    cax = fig.add_axes(
        [
            0.82,    # posicao x (entre 0.0 e 1.0)
            0.18,    # posicao y (entre 0.0 e 1.0)
            0.03,    # largura x
            0.40,    # altura y
        ]
    )

    sm = plt.cm.ScalarMappable(
        cmap="viridis",                                       
        norm=plt.Normalize(
            vmin=map_br_aux["CASO"].min(),  
            vmax=map_br_aux["CASO"].max(),  
        ),
    )
    
    fig.colorbar(
    sm,
    cax=cax)
    
    axs.set_title(
    "Cases residing abroad from 2007-2022",
    fontdict={"fontsize": 15},
    )
    axs.axis("off") 
    fig.savefig('../Figs/res_cases.pdf')
```
