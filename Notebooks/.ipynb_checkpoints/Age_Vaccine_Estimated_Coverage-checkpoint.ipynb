```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
```

```python
sys.path.append('..')
```

## National cases

```python
df_national = pd.read_csv('../Data/analyzed/age_vaccine_national.csv')
```

### Histogram

```python
min_age = min(df_national['NU_IDADE_N'])
max_age = max(df_national['NU_IDADE_N'])
plt.hist(df_national['NU_IDADE_N'], edgecolor = 'black', bins=np.arange(min_age,max_age+1,1))
plt.xticks(np.arange(min_age,max_age+1,5), rotation = 45)
plt.title('Case Age - National', fontsize = 14)
plt.xlabel('Age', fontsize = 12)
plt.ylabel('Number of Cases', fontsize = 12)
plt.show()
```

```python
min_age = min(df_national['NU_IDADE_N'])
max_age = max(df_national['NU_IDADE_N'])
plt.hist(df_national['NU_IDADE_N'], edgecolor = 'black', bins=np.arange(min_age,max_age+1,1))
plt.xticks(np.arange(min_age,max_age+1,1), rotation = 45)
plt.title('Case Age - National', fontsize = 14)
plt.xlim((15,35))
plt.ylim((0,2500))
plt.xlabel('Age', fontsize = 12)
plt.ylabel('Number of Cases', fontsize = 12)
plt.show()
```

### Discriminating by vaccination status

```python
df_national = df_national.drop(columns = ['Unnamed: 0'])
```

```python
df_national = df_national.dropna()
```

```python
x_vac = list(df_national[df_national['CS_VACINA'] == 1]['NU_IDADE_N'])
x_nvac = list(df_national[df_national['CS_VACINA'] == 2]['NU_IDADE_N'])
x_ign = list(df_national[df_national['CS_VACINA'] == 9]['NU_IDADE_N'])
```

```python
colors=['blue', 'green', 'orange']
names=['VAC', 'NVAC', 'IGN']
```

```python
plt.hist([x_vac, x_nvac, x_ign], color=colors, label=names, bins=np.arange(0,max_age+1,1), edgecolor = 'black')
plt.xticks(np.arange(min_age,max_age+1,1), rotation = 45)
plt.xlim((0,5))
plt.title('Cases by Age and Vaccination Status (1-5 years)')
plt.xlabel('Age', fontsize = 12)
plt.ylabel('Number of Cases', fontsize = 12)
plt.legend()
plt.show()
```

```python
plt.hist([x_vac, x_nvac, x_ign], color=colors, label=names, bins=np.arange(0,max_age+1,1), edgecolor = 'black')
plt.xticks(np.arange(min_age,max_age+1,1), rotation = 45)
plt.xlim((15,35))
plt.ylim((0,1100))
plt.title('Cases by Age and Vaccination Status (15-35 years)')
plt.xlabel('Age', fontsize = 12)
plt.ylabel('Number of Cases', fontsize = 12)
plt.legend()
plt.show()
```
