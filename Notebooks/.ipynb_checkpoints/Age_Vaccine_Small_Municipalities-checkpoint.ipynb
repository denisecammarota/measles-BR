```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
```

```python
sys.path.append('..')
```

## What are the municipalities with the least amount of cases?

```python
df_cases = pd.read_csv('../Data/analyzed/age_vaccine_municipalities.csv')
```

```python
df_cases['CASE'] = 1
```

```python
df_cases_mun = df_cases.groupby(by = ['ID_MN_RESI'])['CASE'].sum()
df_cases_mun = df_cases_mun.reset_index()
```

### Cases per 100k inhabitants

```python
mun_path = '../Data/municipalities.csv'
mun_file = pd.read_csv(mun_path)
mun_file['codigo'] = mun_file['municipio'].astype(str).str[:6].astype(int)
mun_file = mun_file[['municipio','name','codigo','pop_21']]
```

```python
df_cases_mun = df_cases_mun.merge(mun_file, how = 'left', left_on = 'ID_MN_RESI', right_on = 'codigo')
```

```python
df_cases_mun['CASES_100'] = (df_cases_mun['CASE']/df_cases_mun['pop_21'])*(10**5)
```

```python
df_cases_mun = df_cases_mun.sort_values(by = 'CASES_100', ascending = False)
```

```python
last_100 = df_cases_mun[-100:]['ID_MN_RESI'].unique()
last_200 = df_cases_mun[-200:]['ID_MN_RESI'].unique()
last_500 = df_cases_mun[-500:]['ID_MN_RESI'].unique()
```

```python
df_cases_mun.sort_values(by = 'CASE', ascending = False)[:20]
```

## What age profile do they demonstrate?

### Last 100

```python
df_100 = df_cases[df_cases['ID_MN_RESI'].isin(last_100)]
```

```python
df_100 = df_100.drop(columns = ['Unnamed: 0','CASE','ID_MN_RESI'])
```

```python
min_age = 0
max_age = int(max(df_100['NU_IDADE_N']))
plt.hist(df_100['NU_IDADE_N'], edgecolor = 'black', bins=np.arange(min_age,max_age+1,1))
plt.xticks(np.arange(min_age,max_age+1,5), rotation = 45)
plt.title('Case Age - 100 municipalities with least cases', fontsize = 14)
plt.xlabel('Age', fontsize = 12)
plt.ylabel('Number of Cases', fontsize = 12)
plt.show()
```

```python
df_100 = df_100.dropna()
```

```python
x_vac = list(df_100[df_100['CS_VACINA'] == 1]['NU_IDADE_N'])
x_nvac = list(df_100[df_100['CS_VACINA'] == 2]['NU_IDADE_N'])
x_ign = list(df_100[df_100['CS_VACINA'] == 9]['NU_IDADE_N'])
```

```python
colors=['blue', 'green', 'orange']
names=['VAC', 'NVAC', 'IGN']
```

```python
plt.hist([x_vac, x_nvac, x_ign], color=colors, label=names, bins=np.arange(0,max_age+1,1), edgecolor = 'black')
plt.xticks(np.arange(min_age,max_age+1,1), rotation = 45)
plt.title('Cases by Age and Vaccination Status - Last 100')
plt.xlim((0,5))
plt.xlabel('Age', fontsize = 12)
plt.ylabel('Number of Cases', fontsize = 12)
plt.legend()
plt.show()
```

```python
plt.hist([x_vac, x_nvac, x_ign], color=colors, label=names, bins=np.arange(0,max_age+1,1), edgecolor = 'black')
plt.xticks(np.arange(min_age,max_age+1,1), rotation = 45)
plt.title('Cases by Age and Vaccination Status - Last 100')
plt.xlim((15,35))
plt.ylim((0,15))
plt.xlabel('Age', fontsize = 12)
plt.ylabel('Number of Cases', fontsize = 12)
plt.legend()
plt.show()
```
