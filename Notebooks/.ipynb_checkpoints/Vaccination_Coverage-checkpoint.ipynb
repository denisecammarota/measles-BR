```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
```

```python
sys.path.append('..')
```

## Vaccination Coverage by Region (FIRST DOSE)

```python
df_1 = pd.read_csv('../Data/CV_TP_1_REG.csv', encoding='cp1252', delimiter = ';', decimal = ',') 
df_2 = pd.read_csv('../Data/CV_TP_2_REG.csv', encoding='cp1252', delimiter = ';', decimal = ',') 
```

```python
df1_regions = df_1['Regi√£o'].to_numpy()[:-1]
df1_years = np.arange(1999,2023)
```

```python
df_1_values = df_1[df_1.keys()[1:-2]]
```

```python
df_1_values = df_1_values.to_numpy()
```

```python
df_1_values = df_1_values[:5,:]
```

```python
df_1_values = df_1_values.T
```

```python
plt.plot(df1_years,df_1_values,'-o', label = df1_regions)
plt.title('Vaccination Coverage of TP Dose 1 in Brazilian Regions')
plt.xlabel('Year')
plt.ylabel('Vaccination Coverage')
plt.legend()
plt.savefig('../Figs/Vac_Cov_D1_REG.pdf')
plt.show()
```

```python
plt.plot(df1_years,df_1_values,'-o', label = df1_regions)
plt.title('Vaccination Coverage of TP Dose 1 in Brazilian Regions')
plt.xlabel('Year')
plt.ylabel('Vaccination Coverage')
plt.ylim((0,130))
plt.legend()
plt.savefig('../Figs/Vac_Cov_D1_REG_LIM.pdf')
plt.show()
```
