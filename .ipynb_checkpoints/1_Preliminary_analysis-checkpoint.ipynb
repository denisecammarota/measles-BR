```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'measles_age_cases.csv'
df_file = pd.read_csv(file_path, delimiter = ';')
```

```python
df_file
```

```python
df_file = df_file.replace('-',0)
```

```python
df_file_total = df_file.loc[32]
```

```python
df_file = df_file.loc[16:31]
```

```python
df_file = df_file.apply(pd.to_numeric)
```

```python
df_file
```

### Total cases per year

```python
plt.plot(df_file['Ano Sintoma(s)'],df_file['Total'],'-o')
plt.xlabel('Year',fontsize = 14)
plt.ylabel('Cases', fontsize = 14)
plt.show()
```

### Age groups distribution of cases in total

```python
age_groups = np.arange(0,13,1)
age_groups_labels = ['<1','1-4','5-9','10-14','15-19','20-29','30-39','40-49','50-59','60-64','65-69','70-79','80+']
cases_age_total = list(df_file_total.to_numpy())
cases_age_total = cases_age_total[1:-1] 
cases_age_total = list(map(int, cases_age_total))

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
plt.plot(age_groups,cases_age_total,'-o')
plt.xlabel('Age Group', fontsize = 12)
plt.ylabel('Cases in Age Group/Total Cases', fontsize = 12)
ax.set_xticks(age_groups)
ax.set_xticklabels(age_groups_labels,rotation=45)
plt.show()
```

```python
age_groups = np.arange(0,13,1)
age_groups_labels = ['<1','1-4','5-9','10-14','15-19','20-29','30-39','40-49','50-59','60-64','65-69','70-79','80+']
cases_age_total = list(df_file_total.to_numpy())
cases_age_total = cases_age_total[1:-1] 
cases_age_total = list(map(int, cases_age_total))
cases_total = sum(cases_age_total)

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
plt.plot(age_groups,np.array(cases_age_total)/cases_total,'-o')
plt.xlabel('Age Group', fontsize = 12)
plt.ylabel('Cases in Age Group/Total Cases', fontsize = 12)
ax.set_xticks(age_groups)
ax.set_xticklabels(age_groups_labels,rotation=45)
plt.show()
```

```python
print('Cases from 2007-2022: ',cases_total)
```

### Age groups distribution of cases by year

```python
age_groups = np.arange(0,13,1)
age_groups_labels = ['<1','1-4','5-9','10-14','15-19','20-29','30-39','40-49','50-59','60-64','65-69','70-79','80+']
cases = df_file.to_numpy()
n_cols = cases.shape[0]
years = np.arange(2007,2023,1)

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
for n_col in range(n_cols):
    cases_aux = cases[n_col,1:-1]
    cases_total  = cases_aux.sum()
    plt.plot(age_groups,cases_aux/cases_total,'-o', label = years[n_col])
plt.xlabel('Age Group', fontsize = 12)
plt.ylabel('Cases in Age Group/Total Cases', fontsize = 12)
ax.set_xticks(age_groups)
ax.set_xticklabels(age_groups_labels,rotation=45)
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
